각각의 '새' 가 "짹"을 print 하는 로직, "날고있습니다" 를 print 하는 로직 총 2개를 만들어주세요.
(새의 종류는 "앵무새", "참새", "비둘기", "닭", "러버덕"와 같고 다 다른 "짹"과 "날고 있습니다"를 만들어주세요)

그리고 input 으로 "새" 종류 1개 이름을 받고 "소리내기" 를 입력하면 "짹" 하고 "날기"를 입력하면 나는 코드 만들어주세요.

예)
새와 행동을 입력하세요: 앵무새,날기

앵무새가 날고 있습니다.

새와 행동을 입력하세요: 앵무새,짹

앵무새가 "앵무" 울음소리를 냈습니다.

```python
# birds 딕셔너리로 짹, 날기 저장
birds = {
    "앵무새": {"짹":"앵무","날기":"날고 있습니다"},
    "참새" : {"짹":"포로롱", "날기":"잠깐 날았습니다"},
    "비둘기" : {"짹":"구구", "날기":"날 수 있을까요?"},
    "닭" : {"짹":"꼬꼬", "날기":"나는게 쉽지 않아보입니다"},
    "러버덕" : {"짹":"꽉" , "날기":"날 수 없습니다"}
}

bird, action = input("새와 행동을 입력하세요:").split(",")

if action == "짹" :
    print(f"{bird}(이)가 '{birds[bird][action]}' 울음소리를 냈습니다 ")
elif action =="날기":
    print(f"{bird}(이)가 {birds[bird][action]}")
else :
    print("잘못된입력입니다")
```
